### Мы создаем прототип функции для создания соводной таблицы, для того, чтобы потом менять параметры и получать данные. В функции прописываем методы для поиска того или иного праметра. В даннном нам Dataset числа представлены через разделитель ",", что мешало работать функйии. Для этого мы в таблице использовали функцию в экселе "=ПРЕДСТАВИТЬ(T2;",";".")". 

```python
def generate_table(df, column_name="МесяцДатыЗаказа"): 
    columns_list = [column_name, 'Выручка', 'Абсолютная маржа', 'Маржа', 'Кол-во чеков', 'Кол-во товаров', 'Кол-во уникальных клиентов', 'Средний чек', 'Среднее кол-во товаров в  чеке', 'Ценность клиента'] 
    unique_values = df[column_name].unique()
    total = pd.DataFrame(columns=columns_list)
    for val in unique_values: 
        list_of_values = [val] 
        filtered_df = df[df[column_name] == val] 
        # Подсчет значений
        profit = int(round(filtered_df['НоваяСуммаСтроки'].sum()))               #выручка 
        list_of_values.append(profit) 
        
        absmargin = int(round(filtered_df['ИспМаржа'].sum()))                    #абсолютная маржа 
        list_of_values.append(absmargin) 
        
        margin = absmargin/profit                                                #маржа
        list_of_values.append(margin) 
        
        df_numb_check = filtered_df['НомерЗаказаНаСайте'].unique()               #кол-во чеков 
        numb_check = df_numb_check.size
        list_of_values.append(numb_check) 
        
        numb_good = filtered_df['МесяцДатыЗаказа'].size                          #кол-во товаров 
        list_of_values.append(numb_good) 
        
        num_of_customers = filtered_df['Телефон_new'].unique().size              #кол-во уникальных клиентов
        list_of_values.append(num_of_customers)

        avarage_bill = int(round(filtered_df.groupby(['НомерЗаказаНаСайте']).mean()['НовСуммаЗаказаНаСайте'].mean())) #средний чек 
        list_of_values.append(avarage_bill) 
        
        avarage_numb_good = int(round(filtered_df.groupby(['НомерЗаказаНаСайте']).count()['НовСуммаЗаказаНаСайте'].mean())) #среднее кол-во товаров в  чеке
        list_of_values.append(avarage_numb_good) 

        value_of_customer = int(round(absmargin /num_of_customers))              #ценность клиента
        list_of_values.append(value_of_customer) 
        
        tmp_df = pd.DataFrame([list_of_values], columns=columns_list) 
        total = total.append(tmp_df, ignore_index=True)
    return total 
```

```python
pivot_table = generate_table(df, column_name="МесяцДатыЗаказа") 
```
